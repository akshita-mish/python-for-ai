```python
# 1. Import the dataset fetcher
from sklearn.datasets import fetch_california_housing
```

```python
df=fetch_california_housing()
df
```

```python
dataset=pd.DataFrame(df.data)
```

```python
#setting a column name into it 
dataset.columns=df.feature_names
dataset
```

```python
dataset.shape # get the shape of data 
```

```python
#Independent feature and dependent feature 
X=dataset
Y=df.target
Y
```

```python
# train test split 
from sklearn.model_selection import train_test_split

# Data ko train aur test mein split kar rahe hain
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# to see x 
print("--- Independent Test Features (X_test) ---")
print(X_test.head())

print("\n--- Dependent Test Target (y_test) ---")
print(Y_test[:5])

```

```python
X_train #to train model 

```

```python
#Standardizing the data 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train=scaler.fit_transform(X_train) # fit _transform is fro training  data 
x_test=scaler.transform(X_test)#for testing data here we will not fit it instead model will do its work 
```

```python
#import LinearRegression 
from sklearn.linear_model import LinearRegression
##Cross validation 
from sklearn.model_selection import cross_val_score
```

```python
regression =LinearRegression()
regression.fit(x_train,y_train)
mse=cross_val_score(regression,x_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
np.mean(mse)
```

```python
#prediction 
reg_pred=regression.predict(x_test)
```
