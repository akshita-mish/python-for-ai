Matrix

```python
matrix = np.array([[1,2,3,4],[5,6,7,8]])
print("matrix",matrix)
```

Tensor

```python
tensor = np.array([[1,2,34],
                   [4,5,6],
                   [3,7,8]])
print(tensor)
```

```python
arr = np.array([[1,2,3],
                [4,5,6]])
print("Shape",arr.shape)
print("Dimensions",arr.ndim)
print("Size",arr.size)
print("data type",arr.dtype)
```

Array Reshaping 

```python
arr= np.arange(6)
print("Original array ",arr)

reshaped=arr.reshape((2,3))
print("reshaped array ",reshaped)

flatten =reshaped.flatten()
print("Flatten array " , flatten)
#it return view instead of copy 
raveled = reshaped.ravel()
print("ravel array",raveled)

#Transpose matrix 
transpose= reshaped.T
print("transpose of a matrix ",transpose)
```
