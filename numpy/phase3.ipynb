Trying to apply numpy concept on data to enhance my skills 


```python
employees = np.array([
    [101, 35000, 25],
    [102, 42000, 28],
    [103, 39000, 30],
    [104, 51000, 35],
    [105, 46000, 29]
])

print("Anlysing this employees data ")
print("Shape of data",employees.shape)
```

```python
#  find the sum of salaries of all the employees
sum_of_salary=np.sum(employees[:,1:2],axis=0)
print(sum_of_salary)
```

```python
# 3 find the minimum salary 
min_salary=np.min(employees[:,1:2],axis=0)
print(min_salary)
```

```python
# 3 find the maximum salary 
max_salary=np.max(employees[:,1:2],axis=0)
print(max_salary)
```

```python
# find the older person among all 
older=np.max(employees[:,2:],axis=0)
print(older)
```

```python
#average of salry 
avg= np.mean(employees[:,1:2],axis=0)
print(avg)
```
