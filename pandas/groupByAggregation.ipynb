```python
import numpy as np 
import pandas as pd 
```

```python



# Creating the practice dataset
data = {
    'OrderID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 111, 112, 113, 114, 115],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Home', 'Clothing', 
                 'Home', 'Electronics', 'Clothing', 'Home', 'Electronics', 
                 'Clothing', 'Home', 'Electronics', 'Clothing', 'Home'],
    'SubCategory': ['Laptop', 'T-Shirt', 'Phone', 'Lamp', 'Jeans', 
                    'Chair', 'Phone', 'T-Shirt', 'Lamp', 'Laptop', 
                    'Jeans', 'Chair', 'Headphones', 'T-Shirt', 'Chair'],
    'Region': ['East', 'West', 'East', 'North', 'West', 
               'North', 'South', 'South', 'West', 'North', 
               'East', 'South', 'West', 'North', 'East'],
    'Sales': [1200, 25, 800, 45, 60, 150, 850, 30, 50, 1300, 55, 120, 150, 35, 140],
    'Quantity': [1, 2, 1, 1, 1, 2, 2, 1, 2, 1, 1, 2, 3, 2, 2],
    'Profit': [200, 5, 150, 10, 15, 35, 180, 8, 12, 250, 12, 25, 30, 7, 30]
}
```

```python
df= pd.DataFrame(data)
df
```

```python
#Group by category and calculate the sum of sales 
cat=df.groupby("Category") # this will give an object so we need a loop 
for i,v in cat :
    print(i)
    print(v)
   
```

```python
#if i need to print the sum of sales 
cat1=df.groupby("Category")["Sales"].sum()
cat1
```

```python
#Group by subcategory and calculate the sum of sales 
subCat=df.groupby("SubCategory")["Sales"].sum()
subCat
```

```python
#Group by multiple column 
cat=df.groupby(["Category","SubCategory"])["Sales"].sum()
cat
```
