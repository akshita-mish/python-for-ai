Merging two dataframes 

```python
students = {
    "Student_ID": [101, 102, 103, 104, 105, 106, 107],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace"],
    "Course": ["CSE", "IT", "ECE", "ME", "CSE", "IT", "ECE"],
    "Age": [20, 21, 22, 23, 20, 21, 22]
}
df_students = pd.DataFrame(students)
df_students
```

```python
marks = {
    "Student_ID": [101, 102, 103, 105, 106, 107, 108],
    "Maths": [88, 76, 91, 82, 79, 95, 70],
    "Science": [90, 72, 94, 85, 80, 98, 68],
    "English": [85, 70, 92, 88, 76, 96, 74]
}
df_marks = pd.DataFrame(marks)
df_marks
```

```python
pd.merge(df_students,df_marks,on="Student_ID",how="inner")
```

```python
pd.merge(df_students,df_marks,on="Student_ID",how="outer") #merges on the basis of column 
```

```python
pd.merge(df_students,df_marks,on="Student_ID",how="left") #merging on the basis of left table 
```

```python
pd.merge(df_students,df_marks,on="Student_ID",how="right")
```

Concatenation of two dataframes

```python
cse_students = {
    "Student_ID": [101, 102, 103],
    "Name": ["Alice", "Bob", "Charlie"],
    "Course": ["CSE", "CSE", "CSE"],
    "Marks": [88, 76, 91]
}

df_cse = pd.DataFrame(cse_students)
```

```python
df_cse
```

```python
it_students = {
    "Student_ID": [104, 105, 106],
    "Name": ["David", "Eva", "Frank"],
    "Course": ["IT", "IT", "IT"],
    "Marks": [65, 82, 79]
}

df_it = pd.DataFrame(it_students)
```

```python
df_it
```

```python
pd.concat([df_it,df_cse]) #this will concatenate based on columns 
```

```python
pd.concat([df_it,df_cse],axis=1) #it will concate rows wise 
```

Joining of dataframes 

```python
import pandas as pd

students = {
    "Student_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Course": ["CSE", "IT", "ECE", "ME", "CSE"]
}

df_students = pd.DataFrame(students).set_index("Student_ID")
```

```python
df_students
```

```python
marks = {
    "Student_ID": [101, 102, 103, 105, 106],
    "Maths": [88, 76, 91, 82, 70],
    "Science": [90, 72, 94, 85, 68]
}

df_marks = pd.DataFrame(marks).set_index("Student_ID")
```

```python
df_marks
```

```python
df_students.join(df_marks)
```
