```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Student_ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Henry", "Ivy", "Jack"],
    "Age": [20, 21, np.nan, 22, 20, 23, 21, np.nan, 22, 20],
    "Gender": ["Female", "Male", "Male", "Male", "Female", "Male", "Female", "Male", "Female", "Male"],
    "Math": [85, 78, 92, 65, 88, np.nan, 95, 70, 90, 82],
    "Science": [90, np.nan, 94, 70, 91, 75, 98, 68, np.nan, 85],
    "English": [88, 81, 90, np.nan, 89, 72, 96, 71, 93, np.nan],
    "Attendance": [95, 89, 97, 85, np.nan, 80, 99, 75, 92, 88],
    "City": ["Delhi", "Mumbai", "Bangalore", np.nan, "Pune", "Delhi", "Mumbai", "Bangalore", np.nan, "Chennai"]
}
```

```python
df = pd.DataFrame(data)
```

find null value 

```python
df.isna()
```

```python
df.isna().sum()
```

Removing missing values (Cleaning the data )

```python
#it works based on rows 
df.dropna() #it removes all the rows that contains null  but just a snapshot 
```

```python
#using threash ( it will give only those row that will have a particular non null columns )
df.dropna(thresh=8)
```

Filling the missing value 

```python
df.fillna(2)
```
